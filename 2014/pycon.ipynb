## Character encoding and Unicode in Python: How to (╯°□°）╯︵ ┻━┻ with dignity

Presented by _Travis Fischer_ and _Esther Nam_

[Video](http://pyvideo.org/video/2625/character-encoding-and-unicode-in-python)
[Slides](http://www.slideshare.net/fischertrav/character-encoding-unicode-how-to-with-dignity-33352863)

This talk didn't address Python 3.

A _character_ is mapped to a _Unicode Code Point_.

A _UTF_ (Unicode Transformation Format) provides a binary encoding of code points.

Three layers:

| Layer   | Description      |
|---------|------------------|
| Display | Glyphs and Fonts |
| Text    | Unicode          |
| Storage | Binary UTF-8     |

- `decode` converts from Bytes to Unicode
- `encode` converts from Unicode to Bytes

Note: the encoding cannot be inferred from the bytestring - it **must** be correctly declared.

Best Practices

### Declare the encoding of the source file

```python
# -*- coding: utf-8 -*-
```

### Know your encodings

Python uses the default system encoding unless you tell it otherwise. Always declare the encoding to use when calling `decode`.

### Use the Unicode sandwich

- Decode as early as possible.
- Unicode everywhere in the middle.
- Encode as late as possible.

### Test your text-related code

- use both text and bytes in tests.
- test with incorrect encodings.

## By Your Bootstraps: Porting Your Application to Python3

Presented by _Tres Seaver_

[Video](http://pyvideo.org/video/2626/by-your-bootstraps-porting-your-application-to-p)
[Slides](https://speakerdeck.com/pycon2014/by-your-bootstraps-porting-your-app-to-python3-by-tres-seaver)

The approach to use (originally, but incorrectly, thought to be untenable) is to straddle Python 2 and Python 3 in a single codebase.

Porting should be seen as an iterative process.

Adding test coverage is critical.

Perceived risks:

- Ports are opportunities for bug injection.
- Fear of breaking working software.

Managing risks:

- Improved testing
- Modernized idioms
- Clarity in text vs. bytes

Porting should be done iteratively, from the bottom up. This is effectively a topological sort on the "depends-on" relation.

Use modern idioms (example: `with open()` context managers.

Clearly distinguish between bytes and text - use the `b''` and `u''` forms for _all_ literals.

"Untested code is where the bugs go to hide." Aim for 100% coverage.

Unit testing is preferable for libraries, functional testing is best for applications.

Assert _contracts_ in the tests, not implementation details.

Use `tox`, and don't forget pypy. Note: coverage measures slows down tests - it may be better to run coverage analysis once, separately.

----

Note: Testing C is harder!

Recommendation is to provide a reference implementation in Python.

## Designing Django's Migrations

Presented by _Andrew Godwin_

[Video](http://pyvideo.org/video/2630/designing-djangos-migrations)
[Slides](https://speakerdeck.com/pycon2014/designing-djangos-migrations-by-andrew-godwin)

Most Django users have been using the `South` package to manage schema migrations. `South` needed to monkey-patch a variety of hacks into Django.

This talk focused on the new Django 1.7 migrations.

"Please, stop using MySQL."

In Django, the schema editor and migrations are kept logically separate.

The schema editor abstracts across databases, working in terms of Django fields and models.

There is a declarative version of migrations available:

```python
class Migration(django.db.migrations.Migration):

    operations = {
        ...
    }
```

Of note, there's nothing else in the usual example, just one class-level data element.

## Designing Poetic APIs

Presented by _Erik Rose_

[Video](http://pyvideo.org/video/2647/designing-poetic-apis)
[Slides](https://speakerdeck.com/pycon2014/designing-poetic-apis-by-erik-rose)

Remark: Don't be an architecture astronaut; the best libraries are extracted, not invented.

An example library used to illustrate a number of points in this talk is [nose-progressive](https://pypi.python.org/pypi/nose-progressive/).

### Consistency

- If we have to keep looking at documentation, it's a problem.

### Brevity

Warning signs:

- Copying and pasting.
- "Why can't I just assume this?"
- Long argument lists.

### Composability

Warning signs:

- Classes with lots of state
- Deep inheritance hierarchies.
- Violations of the Law of Demeter (parent class properties and behaviors should be respected by subclass properties and behaviors).
- Mocking in tests.

"Testable code is decoupled code."

### Plain data

Warning signs:

- Users take output and change it
- Instantiating an object just to pass it to another.

### Grooviness

- Fail shallowly. :D
- RAII

Warning signs:

- invariants that aren't
- lack of a clear starting point.
- long, complicated documentation.



## The Sorry State of SSL

Presented by _Hynek Schlawack_

[Video](http://pyvideo.org/video/2651/the-sorry-state-of-ssl)
[Slides](https://speakerdeck.com/pycon2014/the-sorry-state-of-ssl-by-hynek-schlawack)

RSA lacks perfect forward secrecy (PFS). This means that once a key is compromised, all messages previously signed with that key and all future messages signed with that key are compromised.

PFS is impossible with Python 2.7.

Moreover, the standard library cryptographic routines are bound to the openssl library linked into Python.

## In Depth PDB

Presented by _Nathan Yergler_

[Video](http://pyvideo.org/video/2673/in-depth-pdb)

PDB is, in part, a post-mortem debugger. It helps you see what was happening before the program died.

Breakpoints can be set at any callable or line number in a file.

_Conditions_ can be set for breaking only in a specific situation.

PDB can be extended using _aliases_ and _breakpoint commands_.

These types of extensions can be stored in the `.pdbrc` file.

## Postgres Performance for Humans

Presented by _Craig Kerstiens_

[Video](http://pyvideo.org/video/2641/postgres-performance-for-humans)
[Slides](https://speakerdeck.com/pycon2014/postgres-performance-for-humans-by-craig-kerstiens)

PostgreSQL has become more of a data _platform_ than a database.

Performance properties for

- OLTP (for e.g. webapps)
- OLAP (for business intelligence / reporting)

Mostly focused on OLTP

Things to look at:

- Cache hit rate --- PostgrSQL is really good at this, aim for 99% hit rate.
- Index hit rate

You can set up shortcuts for psql in `.psqlrc`

### How data is retrieved

| Sequential Scanning | Index Scanning |
|----|----|
| Large reports | Small queries |
| OLAP | OLTP |

PostgreSQL 9.2+ supports some non-disc access to indices

Use EXPLAIN while analyzing query performance

He gets common queries to < 10 ms, usually 1.5 - 2 ms / query.

Also, `pg_stat_statements` is useful.

### Indexes

| Type | Notes |
|----|----|
| BTree | Usually what you want |
| GIN | Multiple values in a column, arrays, hstores |
| GIST | Full text search, shapes |
| KNN | Similarity |
| SP-GIST | (good for phone numbers) |
| VODKA | (coming soon) |

Supports _conditional_ and _functional_ indexes

Allows concurrent creation of indexes, without holding a lock on the table. A lock is taken for a couple of milliseconds.

hstore (key/value pair) and JSON supported; binary JSON format in the database starting with PostgreSQL 9.4.

### Pooling

SQLAlchemy does this well at the application / framework level

### Backups

- Logical -- `pg_dump` -- human readable
- Physical -- the bytes on disk
